```python
import pandas as pd
import numpy as np

matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

print("Data loaded successfully")
print(f"Matches: {matches.shape}")
print(f"Deliveries: {deliveries.shape}")
```

```python
# Fill missing city with venue name (best guess)
matches['city'] = matches['city'].fillna('Unknown')

print("Missing cities filled")
print(f"Remaining missing cities: {matches['city'].isnull().sum()}")
```

```python
# Method is NaN when DLS not applied — fill with 'Normal'
matches['method'] = matches['method'].fillna('Normal')

print("Method column fixed")
print(matches['method'].value_counts())
```

```python
# result_margin NaN means match had no result (tie/no result)
matches['result_margin'] = matches['result_margin'].fillna(0)

print("Result margin fixed")
```

```python
# Some teams changed names over the years
# Let's see all unique team names first
print("All team names:")
all_teams = pd.concat([matches['team1'], matches['team2']]).unique()
for team in sorted(all_teams):
    print(f"  {team}")
```

```python
# Standardize old team names to current names
team_name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Pune Warriors': 'Pune Warriors',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Kings XI Punjab': 'Punjab Kings',
}

matches['team1'] = matches['team1'].replace(team_name_map)
matches['team2'] = matches['team2'].replace(team_name_map)
matches['toss_winner'] = matches['toss_winner'].replace(team_name_map)
matches['winner'] = matches['winner'].replace(team_name_map)

deliveries['batting_team'] = deliveries['batting_team'].replace(team_name_map)
deliveries['bowling_team'] = deliveries['bowling_team'].replace(team_name_map)

print("Team names standardized")
print(f"Total unique teams now: {pd.concat([matches['team1'], matches['team2']]).nunique()}")

team_name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
}
```

```python
# Convert date to proper datetime format
matches['date'] = pd.to_datetime(matches['date'])
matches['year'] = matches['date'].dt.year
matches['month'] = matches['date'].dt.month

print("Date column converted")
print(matches[['date', 'year', 'month']].head())
```

```python
# Save cleaned versions
matches.to_csv('matches_clean.csv', index=False)
deliveries.to_csv('deliveries_clean.csv', index=False)

print("Clean data saved successfully")
print(f"matches_clean.csv: {matches.shape}")
print(f"deliveries_clean.csv: {deliveries.shape}")
```

```python
# Verify no critical missing values remain
print("Final missing value check:")
print(matches.isnull().sum())
print()
print("Data types:")
print(matches.dtypes)
```
